```python
title_col = []
title_data = titles_data['title']


for t in title_data:
    if t == "Senior Engineer":
        title_col.append(1)
    elif t == "Staff":
        title_col.append(2)
    elif t == "Engineer":
        title_col.append(3)
    elif t == "Senior Staff":
        title_col.append(4)
    elif t == "Assistant Engineer":
        title_col.append(5)
    elif t == "Technique Leader":
        title_col.append(6)
    elif t == "Manager":
        title_col.append(7)
        
titles_data['title_id'] = title_col

titles_data['id'] = range(1, len(titles_data) + 1)

titles_data = titles_data[["emp_no", "title", "from_date", "to_date", "title_id", "id"]] 
```
